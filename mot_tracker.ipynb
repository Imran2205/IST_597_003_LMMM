```python
found = './YOLOX_outputs/yolox_x_mix_det/track_vis/2024_03_08_19_06_24/output_480.mp4'
```

```python
from moviepy.editor import VideoFileClip


def convert_mp4_to_webm(source_file, target_file, fps=16):
    # Load the source video
    clip = VideoFileClip(source_file)
    
    # If fps is specified, set the clip's fps to the new value
    if fps is not None:
        clip = clip.set_fps(fps)
    
    # Write the clip to a new file in WebM format
    clip.write_videofile(target_file, codec='libvpx', fps=fps)


fps = 16

target_webm = 'tracked_out.webm'
convert_mp4_to_webm(found, target_webm, fps)
```

```python
from IPython.display import HTML
from base64 import b64encode

mp4 = open(target_webm, "rb").read()
data_url = "data:video/webm;base64," + b64encode(mp4).decode()
HTML(f"""
<video width=600 controls>
      <source src="{data_url}" type="video/webm">
</video>
""")
```
